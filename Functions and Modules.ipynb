# 3. Variable Length Arguments

```python
def sumofnums(*n):
    result=0
    for x in n:
        result=result+x
    print("The Sum : ",result)
    
sumofnums()
sumofnums(1)
sumofnums(1,2)
sumofnums(1,2,3)
sumofnums(1,2,3,4)
sumofnums(1,2,3,4,5,6,7)
sumofnums(1,2,3,4,6,7,7,8,9,90,0)
```

```python
import random
import string

def generate_captcha(size=6):
    # Use only uppercase letters + digits to avoid case confusion
    chars = string.ascii_uppercase + string.digits
    return ''.join(random.choice(chars) for _ in range(size))

def generate_otp(size=6):
    # Digits only
    return ''.join(random.choice(string.digits) for _ in range(size))

# --- CAPTCHA ---
captcha = generate_captcha()
print("CAPTCHA:", captcha)
user_captcha = input("Enter CAPTCHA: ").strip().upper()

if user_captcha != captcha:
    print("Wrong CAPTCHA. Exiting.")
    exit()

print("CAPTCHA verified.")

# --- OTP ---
otp = generate_otp()
print("OTP:", otp)  # For demo only. In real apps, send via SMS/Email and don't print it.
user_otp = input("Enter OTP: ").strip()

if user_otp == otp:
    print("OTP verified.")
else:
    print("Wrong OTP.")
```
