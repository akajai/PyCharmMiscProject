# Matplotlib

It is a library with collection of commands used for plotting various plots or figures with respect to the given data.
pyplot function in matplotlib library defines certain methods for plotting 

A Matplotlib figure can be categorized into several parts as below:

<b>Figure:</b> It is a whole figure which may contain one or more than one axes (plots). You can think of a Figure as a canvas which contains plots.

<b>Axes:</b> It is what we generally think of as a plot. A Figure can contain many Axes. It contains two or three (in the case of 3D) Axis objects. Each Axes has a title, an x-label and a y-label.

<b>Axis:</b> They are the number line like objects and take care of generating the graph limits.

<b>Artist: </b>Everything which one can see on the figure is an artist like Text objects, Line2D objects, collection objects. Most Artists are tied to Axes.



1. <b>Bar Plot:</b> (Simple Bar, Multiple Bar, Stacked Bar, Percentage Bar) 
2. <b>Pie diagram</b>
3. <b>Histogram</b>
4. <b>Line chart</b>
5. <b>Multiple line chart</b>
6. <b>Scatterplot</b>
7. <b>box plot</b>
8. <b>Frequency density curve</b>

## Magic notebook command

Magic commands are lines in a code cell that starts with `%`. There are three magic cells for matplotlib and that can be called before importing the library or when you want to control the display of a plot in a cell

* `%matplotlib inline` is used for printing static plot images
* `%matplotlib notebook` is used for print interactive plots
* `%matplotlib agg` is like inline but doesn't print explicitly the plots

1. plot(x-axis values, y-axis values) — plots a simple line graph with x-axis values against y-axis values
2. show() — displays the graph
3. title(“string”) — set the title of the plot as specified by the string
4. xlabel(“string”) — set the label for x-axis as specified by the string
5. ylabel(“string”) — set the label for y-axis as specified by the string
6. figure() — used to control a figure level attributes
7. subplot(nrows, ncols, index) — Add a subplot to the current figure
8. suptitle(“string”) — It adds a common title to the figure specified by the string
9. subplots(nrows, ncols, figsize) — a convenient way to create subplots, in a single call. It returns a tuple of a figure and number of axes.
10. set_title(“string”) — an axes level method used to set the title of subplots in a figure
11. bar(categorical variables, values, color) — used to create vertical bar graphs
12. barh(categorical variables, values, color) — used to create horizontal bar graphs
13. legend(loc) — used to make legend of the graph
14. xticks(index, categorical variables) — Get or set the current tick locations and labels of the x-axis
15. pie(value, categorical variables) — used to create a pie chart
16. hist(values, number of bins) — used to create a histogram
17. xlim(start value, end value) — used to set the limit of values of the x-axis
18. ylim(start value, end value) — used to set the limit of values of the y-axis
19. scatter(x-axis values, y-axis values) — plots a scatter plot with x-axis values against y-axis values
20. axes() — adds an axes to the current figure
21. set_xlabel(“string”) — axes level method used to set the x-label of the plot specified as a string
22. set_ylabel(“string”) — axes level method used to set the y-label of the plot specified as a string
23. scatter3D(x-axis values, y-axis values) — plots a three-dimensional scatter plot with x-axis values against y-axis values
24. plot3D(x-axis values, y-axis values) — plots a three-dimensional line graph with x-axis values against y-axis values

    The lower and upper range of the bins. Lower and upper outliers
    are ignored. If not provided, *range* is ``(x.min(), x.max())``.
    Range has no effect if *bins* is a sequence.

cumulative : boolean, optional
    If ``True``, then a histogram is computed where each bin gives the
    counts in that bin plus all bins for smaller values. The last bin
    gives the total number of datapoints. If *cumulative* evaluates to less than 0
    (e.g., -1), the direction of accumulation is reversed.
    Default is ``False``

bottom : array_like, scalar, or None
    Location of the bottom baseline of each bin.  If a scalar,
    the base line for each bin is shifted by the same amount.
    If an array, each bin is shifted independently and the length
    of bottom must match the number of bins.  If None, defaults to 0.

histtype : {'bar', 'barstacked', 'step',  'stepfilled'}, optional
    The type of histogram to draw.
    - 'bar' is a traditional bar-type histogram.  If multiple data
      are given the bars are arranged side by side.
    - 'barstacked' is a bar-type histogram where multiple
      data are stacked on top of each other.
    - 'step' generates a lineplot that is by default
      unfilled.
    - 'stepfilled' generates a lineplot that is by default
      filled.
      Default is 'bar'

align : {'left', 'mid', 'right'}, optional
    Controls how the histogram is plotted.
    - 'left': bars are centered on the left bin edges.
    - 'mid': bars are centered between the bin edges.
    - 'right': bars are centered on the right bin edges.
Default is 'mid'

orientation : {'horizontal', 'vertical'}, optional
    If 'horizontal', `~matplotlib.pyplot.barh` will be used for
    bar-type histograms and the *bottom* kwarg will be the left edges.

color : color or array_like of colors or None, optional

label : string or None, optional
    String, or sequence of strings to match multiple datasets.  Bar
    charts yield multiple patches per dataset, but only the first gets
    the label, so that the legend command will work as expected.
    default is ``None``
    
stacked : boolean, optional
    If ``True``, multiple data are stacked on top of each other If
    ``False`` multiple data are arranged side by side if histtype is
    'bar' or on top of each other if histtype is 'step'

    Default is ``False``

# Box Plot
In order to be an outlier, the data value must be:
-larger than Q3 by at least 1.5 times the interquartile range (IQR=>q3-q1), or
-smaller than Q1 by at least 1.5 times the IQR.

Signature: plt.boxplot(x, notch=None, sym=None, vert=None, whis=None, positions=None, widths=None, patch_artist=None, bootstrap=None, usermedians=None, conf_intervals=None, meanline=None, showmeans=None, showcaps=None, showbox=None, showfliers=None, boxprops=None, labels=None, flierprops=None, medianprops=None, meanprops=None, capprops=None, whiskerprops=None, manage_xticks=True, autorange=False, zorder=None, hold=None, data=None)
Docstring:
Make a box and whisker plot.

Make a box and whisker plot for each column of ``x`` or each
vector in sequence ``x``.  The box extends from the lower to
upper quartile values of the data, with a line at the median.
The whiskers extend from the box to show the range of the
data.  Flier points are those past the end of the whiskers.

sym : str, optional: The default symbol for flier points. 

vert : bool, optional (True):If `True` (default), makes the boxes vertical. If `False`,everything is drawn horizontally.

labels : sequence, optional:Labels for each dataset. Length must be compatible with dimensions  of ``x``.
