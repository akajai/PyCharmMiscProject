```python
import math

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from networkx.algorithms.bipartite.basic import color
import networkx as nx
import tensorflow as ts
from tensorflow import  keras
x=np.linspace(-10,10,200)
#print(x)
y=x*2
y1=x**3
y2=np.power(x,2)
y3=np.power(x,.5)
y4=np.power(2,x)
y5=np.sin(x)
plt.figure(figsize=(8, 5))
plt.plot(x,y,label="x" ,color='r')
plt.plot(x,y1,label="y1=cubic",color='g')
plt.plot(x,y2,label="y2",color='b')
plt.plot(x,y3,label="y3",color='c')
plt.plot(x,y4,label="y4",color='r')
plt.plot(x,y5,label="y5",color='c')
plt.legend()
plt.xlim(min(x),max(x))
plt.ylim(min(x),max(x))
plt.grid()
plt.show()
```
