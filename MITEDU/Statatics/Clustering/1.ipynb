```python

import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import plotly.express as px
import scipy.stats as stats
import scipy.cluster.hierarchy as sch
import sklearn.cluster as cluster
from sklearn.cluster import KMeans

df=pd.read_excel('P1-StartupExpansion.xlsx')
df.head()
df1=df.select_dtypes(include=['number'])
df1.head()
```

```python



    kmean=KMeans(random_state=42,n_clusters=3)
df2=df1.iloc[:,[0,2]]

kmean.fit(df1)
print(df.shape)
print(df1.shape)
print(df2.shape)
labels=kmean.labels_
centroids=kmean.cluster_centers_
print(labels.shape)
print(centroids.shape)
print(centroids)


print(labels)
X = df1.values   # your 2 features
print(X)
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Plot data points
ax.scatter(
    X[:,0],
    X[:,1],
    X[:,2],
    c=labels,
    cmap='viridis',
    alpha=0.6
)
ax.scatter(
    centroids[:,0],   # Feature 1
    centroids[:,1],   # Feature 2
    centroids[:,2],   # Feature 3
    s=200,
    marker='X'
)

ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Feature 3")

ax.set_title("3D Plot of K-Means Centroids")

plt.show()
# Plot centroids (single color)
#plt.scatter(    centroids[:,0],    centroids[:,1],    marker='X',    s=200,    c='red')

#plt.show()
```
