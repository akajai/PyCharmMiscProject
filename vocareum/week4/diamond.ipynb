```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv('diamonds_practice_assignment.csv')
df
```

```python
df1=df.fillna(method='ffill')
df1
```

```python
df['ppc'] = df['price']/df['carat'].replace(0,np.nan)
df.groupby('cut')['ppc'].median().sort_values(ascending=False)
```
