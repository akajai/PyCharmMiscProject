```python

a = 10
b = 3

print("Arithmetic Operators:")
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)    # Floor division
print("a % b =", a % b)      # Modulus
print("a ** b =", a ** b)    # Power
print()



print("Comparison Operators:")
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
print()


# -----------------------------
x = True
y = False

print("Logical Operators:")
print("x and y:", x and y)
print("x OR y:", x or y)
print("NOT x:", not x)
print()


num = 5
print("Assignment Operators:")

num += 3   # num = num + 3
print("num += 3 →", num)

num -= 2
print("num -= 2 →", num)

num *= 4
print("num *= 4 →", num)

num /= 3
print("num /= 3 →", num)

num //= 2
print("num //= 2 →", num)

num %= 3
print("num %= 3 →", num)

num **= 2
print("num **= 2 →", num)
print()


# -----------------------------
p = 6   # 110
q = 3   # 011

print("Bitwise Operators:")
print("p & q (AND) =", p & q)    # 010 -> 2
print("p | q (OR)  =", p | q)    # 111 -> 7
print("p ^ q (XOR) =", p ^ q)    # 101 -> 5
print("~p (NOT)    =", ~p)       # -(p+1)
print("p << 1      =", p << 1)   # Left shift
print("p >> 1      =", p >> 1)   # Right shift
print()



m = [1, 2, 3]
n = m
k = [1, 2, 3]

print("Identity Operators:")
print("m is n:", m is n)         # Same object in memory
print("m is k:", m is k)         # Different objects
print("m is not k:", m is not k)
print()
```
