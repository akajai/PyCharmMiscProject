# WAP to find number of leap years between 1990 to 2020 with the years 

```python
import calendar as cal
lower=int(input("Enter the starting year :  "))
upper=int(input("Enter the ending year :   "))
def leapyear(lower,upper):
    res=[]
    count=0
    for i in range(lower,upper):
        if cal.isleap(i)==True:
            count+=1
            res.append(i)
        else:
            continue
    return count, res
count,res=leapyear(lower,upper)
print("No of Leap year in given range are : ", count)
print(" Leap year in given range are : ", res)
```
