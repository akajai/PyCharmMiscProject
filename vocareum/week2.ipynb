```python
from exceptiongroup import catch

students= {}
print('Enter "done" to end the program.')
print('1 to Add')
print('2 to update')
print('3 to delete')
print('4 to print')
print('5 or any other  no  exit the loop')
while True:

    option=int(input('Enter option'))
    if option==1:
        name = input("Enter student name: ")
        mark = int(input("Enter mark: "))
        students[name] = mark
    elif option==2:
        name = input("Enter student name: ")
        mark = int(input("Enter mark: "))
        if  students.get(name):
            students[name] = mark
    elif option==3:
        name = input("Enter student name: ")
        if students.get(name):
            del students[name]
    elif option==4:
        print(students)
    else:
        break
```

```python
#FIBANOCI SERIES
def fib(n):
    if n <= 1:
        return n
    return fib(n - 1) + fib(n - 2)

num = int(input("Enter number of terms: "))

print("Fibonacci Series:")
for i in range(num):
    print(fib(i), end=" ")
```

```python
a, b = 0, 1

print("Fibonacci series up to 20:")

while a <= 20:
    print(a, end=" ")
    a, b = b, a + b
```

```python
import math

def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    return math.factorial(n)



try:
    num = int(input("Enter a number: "))

    print("Factorial:", factorial(num))
except ValueError as e:
    print("Error:", e)
finally:
    print("Goodbye!")
```
