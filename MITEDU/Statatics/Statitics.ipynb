```python
import numpy as np
import matplotlib.pyplot as plt

# Fix seed for reproducibility
np.random.seed(42)

# Same mean, different standard deviations
data_low_sd  = np.random.normal(loc=50, scale=5, size=1000)   # Low spread
data_high_sd = np.random.normal(loc=50, scale=20, size=1000)  # High spread

print(data_high_sd)
print(data_low_sd)
# Calculate mean and std
mean1, std1 = np.mean(data_low_sd), np.std(data_low_sd)
mean2, std2 = np.mean(data_high_sd), np.std(data_high_sd)

# Plot
plt.figure(figsize=(12,6))
plt.hist(data_low_sd, bins=40, density=True, alpha=0.6, label=f'Low SD (σ={std1:.2f})')
plt.hist(data_high_sd, bins=40, density=True, alpha=0.6, label=f'High SD (σ={std2:.2f})')

plt.axvline(mean1, color='black', linestyle='--', label='Mean')
plt.title('Standard Deviation: Spread Around the Mean')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
